```python
#standerizing data
ss= StandardScaler()
# Fit and transform the data using the StandardScaler
ss_x=ss.fit_transform(x)
ss_x
```

```python
# Principal Component Analysis (PCA) is performed on the standardized data (ss_x) to reduce dimensionality.
pca=PCA()
# The fit_transform method fits the PCA model to the standardized data and transforms it to principal components.
pc= pca.fit_transform(ss_x)
pc
```

```python
pca=PCA(n_components=13)
pc= pca.fit_transform(ss_x)
pc
```

```python
# The amount of variance that each PCA explains
var=pca.explained_variance_ratio_
var
```

```python
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python

import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
x=data2
```

```python
from sklearn.preprocessing import MinMaxScaler
# Creating an instance of MinMaxScaler
MM=MinMaxScaler()
# Transforming the data 'x' using MinMaxScaler to scale the features between 0 and 1
MM_x=MM.fit_transform(x)
```

```python
#dendrogram
plt.figure(figsize=(10,7))
plt.title('Dendrogram')
dend=sch.dendrogram(sch.linkage(MM_x,method='complete'))
```

```python
# Hierarchical Agglomerative Clustering (HC) is performed on the standardized data (MM_x).
hc=AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='single')
# The fit_predict method fits the HC model to the standardized data and assigns cluster labels to each data point.
y=hc.fit_predict(MM_x)
y
```

```python
data2['hc_cluster']=hc.labels_
data2
```

```python
#k-means clustering
from sklearn.cluster import KMeans
KMeans()
kmeans=KMeans(n_clusters=5)
```

```python
kmeans=kmeans.fit(MM_x)
```

```python
labels=kmeans.predict(MM_x)
```

```python
kmeans.inertia_
# Initialize an empty list to store inertia values for different cluster numbers
clust=[]
for i in range(1,11):
   kmeans = KMeans(n_clusters=i,random_state=0)
   kmeans.fit(MM_x)
   clust.append(kmeans.inertia_)
   clust
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11),clust)
plt.title('Elbow Method')
plt.xlabel('Number of cluster')
plt.ylabel('inertial values')
plt.show()
```

```python
model= KMeans(n_clusters=3)
model.fit(MM_x)
model.labels_
```

```python
data3=pd.concat([data2,pd.DataFrame(model.labels_,columns=['KClusters'])],axis=1)
data3
```

```python
data3.iloc[:,0:5].groupby(data3.KClusters).mean()
```

```python
# We get 3 clusters from hierarchical & KMeans clustering and have same number of clusters with the original data.
```
