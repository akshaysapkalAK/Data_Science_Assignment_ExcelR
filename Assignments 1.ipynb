```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Load the dataset
ak = pd.read_csv(r"C:\dfcity\Data science\assingment DS\Ds1\Cars.csv")

# Extract the MPG data
MPG = ak['MPG']

# Create a probability plot (Q-Q plot)
plt.figure(figsize=(8, 6))
stats.probplot(MPG, dist="norm", plot=plt)
plt.title("Q-Q Plot for MPG")
plt.show()

# Perform a normality test (Shapiro-Wilk test)
stat, p_value = stats.shapiro(MPG)

print("Shapiro-Wilk test statistic:", stat)
print("p-value:", p_value)

# Check the p-value to determine normality
alpha = 0.05
if p_value > alpha:
    print("The data follows a normal distribution (p > 0.05)")
else:
    print("The data does not follow a normal distribution (p <= 0.05)")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Load the dataset
ak = pd.read_csv(r"C:\dfcity\Data science\assingment DS\Ds1\wc-at.csv")

# Extract the AT and Waist data
AT = ak['AT']
Waist = ak['Waist']

# Create Q-Q plots for AT and Waist
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
stats.probplot(AT, dist="norm", plot=plt)
plt.title("Q-Q Plot for Adipose Tissue (AT)")

plt.subplot(1, 2, 2)
stats.probplot(Waist, dist="norm", plot=plt)
plt.title("Q-Q Plot for Waist Circumference (Waist)")

plt.tight_layout()
plt.show()

# Perform normality tests (Shapiro-Wilk) for AT and Waist
stat_AT, p_value_AT = stats.shapiro(AT)
stat_Waist, p_value_Waist = stats.shapiro(Waist)

print("Shapiro-Wilk test statistic for AT:", stat_AT)
print("p-value for AT:", p_value_AT)
print("Shapiro-Wilk test statistic for Waist:", stat_Waist)
print("p-value for Waist:", p_value_Waist)

# Check p-values to determine normality
alpha = 0.05
if p_value_AT > alpha:
    print("Adipose Tissue (AT) follows a normal distribution (p > 0.05)")
else:
    print("Adipose Tissue (AT) does not follow a normal distribution (p <= 0.05)")

if p_value_Waist > alpha:
    print("Waist Circumference (Waist) follows a normal distribution (p > 0.05)")
else:
    print("Waist Circumference (Waist) does not follow a normal distribution (p <= 0.05)")
```

```python
import scipy.stats as stats

sample_mean = 260
population_mean = 270
sample_std = 90
sample_size = 18
df = sample_size - 1

# Calculate the t-score
t_score = (sample_mean - population_mean) / (sample_std / (sample_size**0.5))

# Calculate the probability using the t-distribution
probability = stats.t.cdf(t_score, df)

print("t-score:", t_score)
print("Probability:", probability)
```
