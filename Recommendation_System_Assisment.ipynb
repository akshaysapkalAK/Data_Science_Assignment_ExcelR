```python
pivot = pd.pivot_table(data=ak, index = 'user_id',columns='title', values='rating')
pivot
```

```python
pivot.fillna(0,inplace=True)
pivot
```

```python
from sklearn.metrics import pairwise_distances

sim = 1 - pairwise_distances(pivot, metric='cosine')
ak_sim = pd.DataFrame(sim)
ak_sim
```

```python
#set index and column name to userid
ak_sim.index = ak['user_id'].unique()
ak_sim.columns = ak['user_id'].unique()
ak_sim
```

```python
ak_sim.iloc[0:10,0:10]
```

```python
np.fill_diagonal(sim,0)
ak_sim.iloc[0:10,0:10]
```

```python
#most similar users
ak_sim.idxmax(axis=1)[0:10]
```
